# Фильтрация данных

Отбор наблюдений по некоторому условию.

Условие:

```
oscar_nominated_movies.year == 2001
```

Выбрать те фильмы, у которых год 2001 и вывести первые 5 из них:

```
oscar_nominated_movies[oscar_nominated_movies.year == 2001].head()
```

Выбрать те фильмы, у которых год 2001 и рейтинг больше или равен 8, вывести первые 5 из них.

Важно: в отличие от python оператор "и" записывается как &, а оператор "или" как |.
Попробуйте поставить and или or и получите ошибку:)

Скобки тоже важно ставить - они задают приоритет операторов

```
oscar_nominated_movies[(oscar_nominated_movies.year == 2001) & (oscar_nominated_movies.rate >= 8)].head()
```

Найдем средний рейтинг всех фильмов и выведем те фильмы, у которых рейтинг больше или равен среднему

```
mean_rate = oscar_nominated_movies["rate"].mean()
mean_rate
```

```
oscar_nominated_movies[oscar_nominated_movies.rate >= mean_rate]
```

## Отбор колонок

Выведем все колонки, в названии которых встречается слово reviews

```
oscar_nominated_movies.filter(like="reviews")
```

Можно явно указать, что будем искать среди названий колонок - парметр axis=1

```
oscar_nominated_movies.filter(like = "reviews", axis=1)
```

Если axis=0, то ищем среди названий строк. У нас строки имеют только номера, поэтому ничего не находим

```
oscar_nominated_movies.filter(like = "reviews", axis=0)
```

# Построение графиков

Pandas может строить графики matplotlib без подключения этой библиотеки.

Вот один из способов и несколько примеров графиков, которые нам понадобятся.

Гистограмма:

Здесь строят гистограмму значений переменной rate. Если название столбца состоит из одного слова или из слов_разделенных_подчеркиванием, то можно вместо oscar_nominated_movies["rate"] писать oscar_nominated_movies.rate

```
oscar_nominated_movies.rate.plot(kind="hist", figsize=(10,6))
```

Ящик с усами, диаграмма размаха:

```
oscar_nominated_movies.gross.plot(kind="box", figsize=(10,6))
```

Диаграмма рассеяния:

```
oscar_nominated_movies.plot(x="rate", y="gross", kind="scatter", figsize=(10,6))
```

```
oscar_nominated_movies.plot(x="rate", y="metascore", kind="scatter", figsize=(10,6))
```

Гексбиновая диаграмма:

```
oscar_nominated_movies.plot(x="rate", y="metascore", kind="hexbin", figsize=(10,6))
```
